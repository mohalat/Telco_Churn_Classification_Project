# **Title: Telcos Churn Classification Project**

The Telco Churn Classification project aims to predict customer churn for a telecom company using machine learning techniques. Customer churn refers to the loss of customers who stop using the service. By analyzing historical customer data, this project builds predictive models that help identify key factors contributing to churn, allowing the company to proactively manage customer retention. The project leverages various classification algorithms, such as Random Forest and Logistic Regression, and evaluates their performance using metrics like accuracy, F1 score, and confusion matrix.

## 1.0 **Business Understanding**

## 1.1 **Introduction**
Telecommunication companies face significant revenue losses when customers discontinue their services, a phenomenon known as customer churn. Customer retention is a critical focus for telecoms as acquiring new customers is often more expensive than retaining existing ones. The goal of this project is to leverage machine learning to predict customer churn, allowing the telecom company to take proactive steps to retain customers and increase profitability. By using classification models, the company aims to predict which customers are most likely to churn and understand the factors influencing churn behavior.

## 1.2 **Project Objective**
The objective of the Telco Churn Classification project is to develop a predictive model that accurately identifies customers likely to churn based on historical data. By leveraging machine learning algorithms, the project aims to uncover key drivers of customer churn, enabling the telecom company to implement targeted retention strategies. The goal is to enhance decision-making, reduce churn rates, and ultimately improve customer retention and business profitability.
### Project Objectives
1. **Build a classification model** to predict customer churn using historical customer data.
2. **Identify key factors** that influence customer churn, such as service usage, customer demographics, billing information, and customer support interactions.
3. **Provide actionable insights** to help the telecom company reduce churn by focusing on customers most at risk.
4. **Evaluate the performance** of different machine learning algorithms and optimize them for better accuracy and interpretability.
5. **Develop data-driven strategies** to enhance customer retention and increase the company's profitability.

## 1.4 **Success Metrics**
1. **Accuracy**: Measure the percentage of correctly predicted churn and non-churn instances.If 75% accuracy it is good; if 80%+ accuracy, it is excellent).
2. **F1 Score**: A balance between precision and recall, important for handling class imbalance (as churn is usually less frequent). Good models have f1 score of 80%+.
3. **Confusion Matrix**: Understand how well the model is predicting true positives, false positives, true negatives, and false negatives.
4. **AUC-ROC Curve**: Assess model performance across various classification thresholds, focusing on its ability to distinguish between churners and non-churners.

```python
def unique_values(df): 
    unique_summary = []  # List to store summary for each column

    for column in df.columns:  # Loop through each column
        values = df[column].unique()  # Get unique values for the column
        count_values = len(values)  # Count the number of unique values

        # Append the summary as a dictionary
        unique_summary.append({
            "Column": column,
            "Unique_Values": values,
            "Count_of_Unique_Values": count_values
        })

    # Convert the list of dictionaries to a DataFrame for better display
    usummary = pd.DataFrame(unique_summary)

    return usummary
```

```python
# DEFINING THE FUNCTION TO VISUALIZE CATEGORICAL DATA

def visual_categorical(train_data, col):
    fig, axes = plt.subplots(2, len(col), figsize=(20, 10))  # 2 rows (countplot + boxplot) instead of 3
    
    for i, column in enumerate(col):

        # Countplot
        count_ax = sns.countplot(x=train_data[column], ax=axes[0, i], color='red')
        axes[0, i].set_title(f"COUNTPLOT OF {column}")

        # Add labels to the countplot
        total = len(train_data)
        for container in count_ax.containers:
            labels = [f"{int(h.get_height())} ({(h.get_height() / total) * 100:.2f}%)" if h.get_height() > 0 else '' for h in container]
            count_ax.bar_label(container, labels=labels, label_type='edge')

        # Boxplot
        sns.boxplot(train_data[column], ax=axes[1, i], color='green')
        axes[1, i].set_title(f"BOXPLOT OF {column}")

        # Add text label for mode
        frequencies = train_data[column].mode()[0]
        axes[1, i].text(0.05, 0.9, f"Mode: {frequencies}", color='blue', ha='center', va='center')


    plt.tight_layout()
    plt.show()
```

```python
# Bivariate Analysis

def cat_vs_target(train_data, all,target):

    n_rows = math.ceil(len(all)/3)

    # Create a figure with n_rows rows and 3 columns
    fig, axes = plt.subplots(n_rows,3, figsize=(20, 6 * n_rows))

    # Flatten the axes array
    axes = axes.flatten()

    # Looping through all columns in 'all'
    for i, column in enumerate(all):

        # Looping through all rows
        ax = axes[i]

        # Using Countplot to visualize the data 
        sns.countplot(data=train_data, x=train_data[column], hue = target, ax=ax, color='green')

        # Adding labels
        ax.set_xlabel(column)
        ax.set_ylabel('count')
        ax.set_title(f"{column} vs {target}")
        ax.set_xticks
        ax.legend(loc='upper right', title="Churn_Status")

        total = len(train_data)
        for container in ax.containers:
            labels = [f"{int(h.get_height())} ({(h.get_height() / total) * 100:.2f}%)" if h.get_height() > 0 else '' for h in container]
            ax.bar_label(container, labels=labels, label_type='edge')

plt.tight_layout()
plt.show()
```

```python
# Displaying Bivariate Analysis on Categorical Columns with Churn

cat_vs_target(train_data, all, "Churn")
```

**Hyperparameter Tuning**

Hyperparameter tuning is performed to crosscheck or verify whether the models (matrix calculated) are performing better or not, despite their high scores. This is done by choosing a model, and changing some of its parameters (adding more parameters) to see which one works best and gives the best prediction. 

In this project the GridSearchCV will be used for the tunning. It makes use of all instances or data points in the parameter grid, unlike others (eg. RandomizedSearch) which chooses a part or group or portion of the data points and work with it. That is why the GridSearchCV is chosen.

```python
# Define the hyperparater grid

from sklearn.model_selection import GridSearchCV 


# Define parameter grid for GridSearchCV
param_grid = {
    'KNN': {
        'model__n_neighbors': [3, 5, 7, 9],
        'model__weights': ['uniform', 'distance'],
        'model__algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
    },
    'LR': {
        'model__C': [0.1, 1.0, 10.0],
        'model__penalty': ['l1', 'l2'],
        'model__solver': ['liblinear']
    },
    'RF': {
        'model__n_estimators': [50, 100, 200],
        'model__max_depth': [None, 10, 20],
        'model__min_samples_split': [2, 5, 10],
        'model__min_samples_leaf': [1, 2, 4]
    },
    'SVC': {
        'model__C': [0.1, 1.0, 10.0],
        'model__gamma': ['scale', 'auto'],
        'model__kernel': ['rbf', 'linear', 'poly']
    },
    'GB': {
        'model__n_estimators': [50, 100, 200],
        'model__learning_rate': [0.05, 0.1, 0.2],
        'model__max_depth': [3, 5, 7]
    },
    'XGB': {
        'model__n_estimators': [50, 100, 200],
        'model__learning_rate': [0.05, 0.1, 0.2],
        'model__max_depth': [3, 5, 7]
    },
    'DT': {
        'model__max_depth': [3, 5, 10],
        'model__min_samples_split': [2, 5, 10]
    }
}
```

```python
from sklearn.model_selection import StratifiedKFold
import pandas as pd

# Create an empty DataFrame to store the results
results_table = pd.DataFrame(columns=['model', 'best_params', 'best_scores'])

# Initialiaze an empty dictionary to hold the GridSearchCV results for each model
grid_searches_tuned = {}

#Initialize a list or dictionary to hold the best models
best_model = {}

# Hyperparameter tuning loop
for model, classifier in models:

    # Create a pipeline for each model
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', classifier),
    ])
    
    # Create a StratifiedKFold object
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Create the GridSearchCV object
    grid_search_tuned = GridSearchCV(
        estimator=pipeline, 
        param_grid=param_grid[model], 
        cv=skf, 
        scoring= ['f1_weighted', 'f1', 'accuracy', 'precision', 'recall'],
        refit = 'f1_weighted',   # True if one scoring is to be used. Refit model with the best score using f1_weighted in this case
        n_jobs = -1
        
        )

    # fit grid search on training data
    grid_search_tuned.fit(X_train, y_train)

    # Store the grid search object in the dictionary
    grid_searches_tuned[model] = grid_search_tuned

    # Add the results to the DataFrame
    best_params = grid_search_tuned.best_params_
    best_score = grid_search_tuned.best_score_

    results_table.loc[len(results_table)] = [model, best_params, best_score]

    # Store the best model in best_models
    best_model[model] = grid_search_tuned.best_estimator_

# Sort the results table by best score
results_table = results_table.sort_values(by='best_scores', ascending=False).reset_index(drop=True)
# Display the results table
results_table
```

**Business Impact Assessment and Documentation of the Model**


- **Importance for Churn Rate**

In the context of predicting customer churn, it is critical to prioritize recall over precision and accuracy due to their practical implications:

- Recall: High recall ensures that the majority of actual churners are correctly identified, minimizing false negatives. This is crucial because each missed churner represents a lost customer, which could have been retained through targeted interventions.


- Recall over Precision: While precision ensures that the predicted churners are mostly actual churners, it is acceptable to have a lower precision if it means reducing the number of missed churners (false negatives). Intervening with non-churners (false positives) might incur some cost but is less damaging than losing customers without any intervention.

- **Choosing the Right Matrix:**

Imbalanced Dataset: For imbalanced datasets, F1 Score and Weighted Accuracy are generally more reliable metrics. The F1 Score focuses on the performance of the minority class by balancing precision and recall, while Weighted Accuracy provides an overall performance measure that accounts for the distribution of classes. Other useful matrices include the ROC AUC Score, which evaluates the ability of the model to distinguish between classes across all levels.

Balanced Dataset: Accuracy, the ratio of correctly predicted instances to total instances, is more meaningful in balanced datasets. Precision and Recall too become equally important without the risk of biasing towards the majority class. The F1 Score continues to be valuable by providing a harmonic mean of Precision and Recall. Macro average is particularly useful for providing a more balanced evaluation when all classes are of equal importance, as it avoids being dominated by the performance on the most prevalent class. Additionally, matrices like Balanced Accuracy, which averages the recall obtained on each class, and the Confusion Matrix, which provides a detailed breakdown of correct and incorrect predictions for each class, are useful.

- **Choosing the Best Model:**

- Best Model: Random Forest (RF)

Reason: Random Forest offers the highest ROC AUC score, indicating superior ability to distinguish between churners and non-churners. Its high recall ensures that most churners are identified, which is vital for retention efforts. Additionally, its excellent F1 score and low log loss demonstrate overall reliability and accuracy, making it the best model for predicting telco churn.

- Second Best Model: GradientBoost (GB)

Reason: GradienBoost performs consistently well across all metrics, with a high ROC AUC score and good recall. This makes it a reliable alternative to Random Forest, ensuring that a substantial portion of churners is identified and that the predictions are reasonably confident.


- **Practical Implications**

High Recall: Ensuring that most churners are identified allows the company to intervene effectively, reducing the churn rate and improving customer retention.

Balanced Metrics: Using models with a good balance of precision, recall, and F1 score ensures that the interventions are both effective and efficient, minimizing unnecessary costs while maximizing retention.

In summary, selecting Random Forest as the primary model, with GradientBoost as a strong alternative, provides a robust strategy for predicting customer churn. These models' high recall and low log loss ensure effective identification and intervention for potential churners, ultimately supporting the company's retention efforts and strategic planning.
